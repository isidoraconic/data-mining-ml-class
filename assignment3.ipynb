### 1. Online Retail Market Basket Analysis

Loading the Online Retail Dataset

```python
import pandas as pd 
pd.set_option("max_colwidth", 150)
```

```python
f = "https://github.com/cs6220/cs6220.spring2019/raw/master/data/Online%20Retail.xlsx"
```

```python
df = pd.read_excel(f)
```

```python
basket = (df[df["Country"] == "United Kingdom"]
.groupby(["InvoiceNo", "Description"])["Quantity"] .sum().unstack().reset_index().fillna(0)
.set_index("InvoiceNo")) # transform transactions into baskets of items
```

```python
basket_sets = basket.applymap(lambda x: 1 if x >=1 else 0) # convert counts to booleans
```

*(c) Use lift for the rule interestingness (metric="lift") and generate rules up to a depth of at least 3 (max len=3) or higher. Generate rules and find at least 5 rules that you find interesting. Comment on your findings and try to reason about these association rules. Decide yourself on the levels of support and confidence used in this analysis.*

From: http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/ 

The lift metric is commonly used to measure how much more often the antecedent and consequent of a rule A->C occur together than we would expect if they were statistically independent. If A and C are independent, the Lift score will be exactly 1.

**NOTE:** Here, I am going to use an even lower min support to try to get some more diverse rules than the ones above.

```python
adult_freq_itemsets = apriori(df, min_support=0.01, use_colnames=True, max_len=3)
adult_freq_itemsets.head()
```

```python
lift_rules = association_rules(adult_freq_itemsets, metric="lift", min_threshold=1)
lift_rules.head(70)
```

```python
lift_rules.sort_values(by=["lift"], ascending=False).head(50)
```

**1. {sex[T. Male], occupation[T. Exec-managerial]} --> {income[T. >50K]}**

This rule has a lift of 2.070635 (where 1 means the events are independent), which suggests that the antecedent and consequent are linked, and do not just happen by chance. It entirely makes sense knowing that men tend to make the most money (when comparing just gender), and the fact that we are also considering an executive managerial position (a generally high earning position), this rule makes sense. 

**2. {occupation[T. Craft-repair]} --> {education[T. HS-grad], sex[T. Male]}**

This rule has a lift of 2.033108, which again implies that the antecedent and consequent are linked, and do not just occur together by chance. Looking at the rule, it makes sense; generally, adults whose highest education level is highschool tend more often to go into a trade career (repair, etc.). Further, seeing as this is a labour intensive job, which are generally more dominated by men, the rule further makes sense.

**3. {occupation[T. Prof-specialty], income[T. >50K]} --> {education[T. Masters]}**

Notice for this rule, that the opposite (reflexive) rule also has the exact same lift, which is 4.259387. This is a high lift value, which implies that being a prof and making more than 50k is tightly linked to having a master's degree. This entirely makes sense, seeing as the vast majority of professors that I have come across, have a master's degree but usually often a PhD as well.

**4. {relationship[T. Unmarried]} --> {marital_status[T. Separated]}**

Notice for this rule, that the opposite (reflexive) rule also has the exact same lift, which is 4.157528. Again, this is a fairly high lift value, which would imply that being unmarried is tightly linked with being separated. This rule also makes sense, as it sheds light on the fact that being separated and currently unmarried tend to go hand in hand (I believe most people aren't quick to get married again as soon as they are separated).

**5. {occupation[T. Farming-fishing], sex[T. Male]} --> {workclass[ Self-emp-not-inc]}**

Notice for this rule, that the opposite (reflexive) rule also has the exact same lift, which is 5.655372. Finally, this rule also makes sense. It implies that there is a tight relationship between men who farm-fish as their occupation, and being self employed by a non-corporation (definition of "Self-emp-not-inc" found here: https://smallbiztrends.com/2013/11/incorporated-versus-unincorporated-self-employment.html). Seeing as the opposite rule also has the same lift, we can reason that this makes sense. As mentioned previously, men are more likely to work in more labour-intensive jobs, and it seems that fish farming is not a big corporate type of job (not corp).
