The comparison between when I used the same city for both testing and evaluating my PM model is stark. We can see that we got a low R-squared values for all the cities in the question above, and in theory, an r-squared value should be between 0 and 1 (the closer to 1, the better the prediction). However, here we can see we got a fiarly large negative value (which would suggest a terrible prediction). From what I have read ([here](http://www.fairlynerdy.com/what-is-r-squared/)), we usually don't see negative R-squared values, because if the prediction is so bad (such that it would yield a negative r-squared value), it is simply better to use the mean of the predictive objective as the prediction. Also, generally, the value cannot be nagtive because we are squaring the values. However, [this article](https://web.maths.unsw.edu.au/~adelle/Garvan/Assays/GoodnessOfFit.html) provides more insight into what a negative r-squared value implies: "Note that it is possible to get a negative R-square for equations that do not contain a constant term. Because R-square is defined as the proportion of variance explained by the fit, if the fit is actually worse than just fitting a horizontal line then R-square is negative. In this case, R-square cannot be interpreted as the square of a correlation. Such situations indicate that a constant term should be added to the model.". In order to decide which city I would want to live in, I thought it would be interesting to consider not only PM (air pollution metric) but also humidity and temperature. Specifically with humidity and temperature, I think it would be most telling to look at "humidex". In Canada, we use humidex to tell people what a temperature actually feels like when humidity is considered. From [wikipedia](https://en.wikipedia.org/wiki/Humidex): "The humidex is an index number used by Canadian meteorologists to describe how hot the weather feels to the average person, by combining the effect of heat and humidity. The term humidex was first coined in 1965. The humidex is a nominally dimensionless quantity based on the dew point." Therefore, I will take a look at the humidex values in each city as well as PM.

To calculate humidex, I will be using the simplified formula from [here](https://www.physlink.com/education/askexperts/ae287.cfm).

````python
def humidex(temp, humi):
    exponent = (7.5*temp/(237.7+temp))
    e = (6.112*10**exponent*humi/100)
    humidex = temp+5/9*(e-10)
    return humidex
````

````python
#Checking if my humidex method is working correctly
print(humidex(30, 70))
````
